```python
games_df_filtered = games_df[["Description", "ImagePath"]].dropna()
texts = games_df_filtered["Description"].tolist()
image_paths = games_df_filtered["ImagePath"].tolist()

print("Ejemplo de descripciones:")
print(texts[:5])
print("\nEjemplo de rutas de imágenes:")
print(image_paths[:5])
```

```python
def load_images_from_urls(image_urls):
    images = []
    for url in image_urls:
        response = requests.get(url, stream=True)
        if response.status_code == 200:
            image = Image.open(response.raw).convert("RGB")
            images.append(image)
        else:
            print(f"Failed to load image from: {url}")
    return images

num_samples = 100
images = load_images_from_urls(image_paths[:num_samples])
```

## Generate Blip embeddings

```python
model_name = "Salesforce/blip-image-captioning-base"
model = BlipModel.from_pretrained(model_name)
processor = BlipProcessor.from_pretrained(model_name)
```

```python
def generate_blip_embeddings(texts, images):
    text_embeds = []
    image_embeds = []
    multimodal_embeds = []

    for text, image in zip(texts, images):
        inputs = processor(
            text=texts,
            images=images,
            return_tensors="pt",
            padding="max_length",
            truncation=True,
            max_length=64
        )

        with torch.no_grad():
            outputs = model(**inputs)

        text_embeds.append(outputs.text_embeds[0])
        image_embeds.append(outputs.image_embeds[0])
        multimodal_embeds.append(outputs.text_embeds[0] + outputs.image_embeds[0])

    return torch.stack(text_embeds), torch.stack(image_embeds), torch.stack(multimodal_embeds)

blip_text_embeds, blip_image_embeds, blip_multimodal_embeds = generate_blip_embeddings(texts[:num_samples], images)

print(f"Embeddings de texto: {blip_text_embeds.shape}")
print(f"Embeddings de imágenes: {blip_image_embeds.shape}")
print(f"Embeddings multimodales: {blip_multimodal_embeds.shape}")
```
